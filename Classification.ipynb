```
import torch
import torch.optim as optim
torch.__version__
```

입력데이터(X)와 출력데이터(Y) 정의

```
x_data = [[1],[2],[3],[4],[5],[6]]
y_data = [[0],[0],[0],[1],[1],[1]]

x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)
```

초기 파라미터(W와 b) 설정

```
W = torch.zeros((1, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

시그모이드 함수를 사용한 논리회귀모델의 가설

```
hypothesis = 1 / (1 + torch.exp(-(x_train.matmul(W) + b)))
```

비용함수(오차) 계산

```
losses = -(y_train * torch.log(hypothesis) +
           (1 - y_train) * torch.log(1 - hypothesis))
cost = losses.mean()
```

경사하강 알고리즘 설정

```
optimizer = optim.SGD([W, b], lr=1)
```

경사하강 알고리즘을 반복해서 동작
반복한 횟수별 Cost값(오차) 출력

```
nb_epochs = 1000
for epoch in range(nb_epochs + 1):

    hypothesis = 1 / (1 + torch.exp(-(x_train.matmul(W) + b)))
    losses = -(y_train * torch.log(hypothesis) +
           (1 - y_train) * torch.log(1 - hypothesis))
    cost = losses.mean()

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

학습한 모델에 대해서 Test하기

```
hypothesis = 1 / (1 + torch.exp(-(x_train.matmul(W) + b)))
print("---Hypothesis(논리회귀모델 결과값)---") ##논리회귀모델 결과값
print(hypothesis)

# 만약 가설 결과값이 0.5보다 크다면, 예측값을 1로 설정
# 그렇지않다면, 예측값을 0으로 설정
torch_tensor = hypothesis >= torch.FloatTensor([0.5])
prediction = torch_tensor.float()
print("---Prediction(예측값)---") ## 예측값
print(prediction) ## 예측값
print("---Ground truth(정답)---") ## 정답
print(y_train) ## 정답

correct_prediction = prediction.float() == y_train  ## 예측값과 정답과 비교

accuracy = correct_prediction.sum().item() / len(correct_prediction)
print('The model has an accuracy of {:2.2f}% for the training set.'.format(accuracy * 100))
```
