입력데이터(X)와 출력데이터(Y) 정의

```
X = [1,2,3]
Y = [1,2,3]
```

첫번째 가설의 비용함수값 계산

```
W = 0.6
b = 2.0

cost = 0
for i in range(len(X)):
  squared_err = ((W * X[i] + b) - Y[i]) ** 2   # 에러 제곱 계산
  cost = cost + squared_err  # 에러 제곱 누적

cost = cost / float(len(X))   # 평균 제곱 에러 계산

print("cost값: %f, W = %f, b = %f" %(cost, W, b))
```

두번째 가설의 비용함수값 계산

```
W = 1.0
b = 0.0

cost = 0
for i in range(len(X)):
  squared_err = ((W * X[i] + b) - Y[i]) ** 2   # 에러 제곱 계산
  cost = cost + squared_err  # 에러 제곱 누적

cost = cost / float(len(X))   # 평균 제곱 에러 계산

print("cost값: %f, W = %f, b = %f" %(cost, W, b))
```
